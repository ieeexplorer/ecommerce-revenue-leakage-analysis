# 03 Visualizations


```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv('../data/cleaned/ecommerce_cleaned.csv', parse_dates=['order_date'])
completed = df[df['order_status'] == 'Completed'].copy()
completed['order_month'] = completed['order_date'].dt.to_period('M').astype(str)
```

```python
# Monthly revenue trend
monthly = completed.groupby('order_month', as_index=False)['revenue'].sum()

plt.figure(figsize=(8, 4))
plt.plot(monthly['order_month'], monthly['revenue'], marker='o')
plt.title('Monthly Revenue Trend')
plt.xlabel('Month')
plt.ylabel('Revenue')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Revenue by category
category = completed.groupby('category', as_index=False)['revenue'].sum().sort_values('revenue', ascending=False)

plt.figure(figsize=(8, 4))
plt.bar(category['category'], category['revenue'])
plt.title('Revenue by Category')
plt.xlabel('Category')
plt.ylabel('Revenue')
plt.tight_layout()
plt.show()
```

```python
# Return rate by category
return_rate = (
    completed.groupby('category', as_index=False)
    .agg(return_rate=('return_flag', 'mean'))
)
return_rate['return_rate_pct'] = return_rate['return_rate'] * 100

plt.figure(figsize=(8, 4))
plt.bar(return_rate['category'], return_rate['return_rate_pct'])
plt.title('Return Rate by Category')
plt.xlabel('Category')
plt.ylabel('Return Rate %')
plt.tight_layout()
plt.show()
```

```python
# Delivery delays by region
delay_by_region = (
    completed.groupby('region', as_index=False)
    .agg(delay_rate=('delay_flag', 'mean'))
)
delay_by_region['delay_rate_pct'] = delay_by_region['delay_rate'] * 100

plt.figure(figsize=(8, 4))
plt.bar(delay_by_region['region'], delay_by_region['delay_rate_pct'])
plt.title('Delivery Delay Rate by Region')
plt.xlabel('Region')
plt.ylabel('Delay Rate %')
plt.tight_layout()
plt.show()
```
